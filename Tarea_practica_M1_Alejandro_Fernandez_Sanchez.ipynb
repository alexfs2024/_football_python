 "1.12. Crear la función para automatizar todo el proceso, todos los procesos anteriores que hemos ido realizando los encapsularemos dentro de una función. La función se llamará ``transform_data()``, el único requisito es que tenga el parámetro de la ruta donde se encuentran los datos. Utilizar la # para nombrar lo que se hace en cada parte del código.\n",

```python
def transform_data(df):
  #Sustituir los valores Nan por 0
  df.fillna(0,inplace=True)
  #Eliminar la columna ``'Rk'`` y ``'matches'``.
  df.drop(["Rk","Matches"] ,axis=1,inplace=True)
  #Solucionar el problema de  los nombres de los jugadores
  df['Player']=df['Player'].apply(lambda name:name.split("\\")[0])
  #Solucionar el problema de la nacionalidad
  df['Nation']=df['Nation'].apply(lambda name:name.split(" ")[1])
  # Solucionar el problema de las posiciones
  df['Pos'] = df['Pos'].apply(lambda x: x[:2])
  #Renombrar las columnas que contiene el valor .1
  df.rename(columns={'Gls.1':'Gls/90','Ast.1':'Ast/90', 'G-PK.1':'G-PK/90','xG.1':'xG/90','xA.1':'xA/90','npxG.1':'npxG/90','npxG+xA.1':'npxG+xA/90'},inplace=True)
  return df
```

3. Bucle for y condicionales\n
3.1. Vamos a realizar un bucle for donde iteraremos por los diferentes valores de la columna ``'Gls'`` para crear una nueva columna que se llamara ``'type_goal_player'``. Dejo a tu elección los valores en base dónde cortar para crear las diferentes categorías. Hay que crear tres valor Poco goleador, goleador, Muy goleador\n"

```python
for i in range(len(df)):

  if df.Gls.values[i] <= 5:
    df.loc[i,'type_goal_player'] = 'Poco goleador'

  elif (df.Gls.values[i]  > 5) & (df.Gls.values[i]  <= 15):
    df.loc[i,'type_goal_player']= 'Goleador'

  else :
    df.loc[i,'type_goal_player'] = 'Muy goleador'

```
